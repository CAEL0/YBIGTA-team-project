```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('wine_of_the_month.csv')
```

```python
df
```

* 첫 column 삭제

```python
df.drop('Unnamed: 0', axis=1, inplace=True)
```

* 가격 정보 정수로 변환

```python
df.가격 = df.가격.apply(lambda x: int(x.replace(',', '').replace('원', '')) if x != '가격정보없음' else np.NaN)
```

* 아로마, 음식 매칭의 \n을 ,로 변경

```python
df.아로마 = df.아로마.apply(lambda x: x.replace('\n', ', ') if isinstance(x, str) else x)
```

```python
df.음식매칭 = df.음식매칭.apply(lambda x: x.replace('\n', ', ') if isinstance(x, str) else x)
```

* 알코올 추출 후 스타일, 등급, 음용온도 제거

```python
df['temp'] = ''
```

```python
for i in range(len(df)):
    for column in ['스타일', '등급', '알코올', '음용온도']:
        data = df.loc[i, column]
        if isinstance(data, str) and (len(data) < 10) and ('%' in data):
            df.loc[i, 'temp'] = data
```

```python
df.알코올 = df.temp
df.drop(['스타일', '등급', '음용온도', 'temp'], axis=1, inplace=True)
```

```python
df['테이스팅 노트'] = df['테이스팅 노트'].map(lambda x: x.replace('테이스팅노트', ''))
```

```python
df.to_csv('wine_of_the_month_final.csv')
```
