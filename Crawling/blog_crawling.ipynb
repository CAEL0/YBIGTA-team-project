```python
wine_lists = []

for i in range(len(wine_list) // 10 + 1):
    wine_lists.append(wine_list[10 * i:10 * (i + 1)])

print(f'{len(wine_list) = }')
print(f'{len(wine_lists) = }')
```

```python
path = 'C:/Users/USER/Desktop/종민/YBIGTA/2021-2 교육세션/0805 크롤링/chromedriver.exe'
driver = webdriver.Chrome(path)
driver.implicitly_wait(5)

url = 'https://www.naver.com/'
driver.get(url)

css = '#query'
search = driver.find_element_by_css_selector(css)
search.send_keys(Keys.RETURN)

css = '#lnb > div.lnb_group > div > ul > li:nth-child(2) > a'
search = driver.find_element_by_css_selector(css)
search.send_keys(Keys.RETURN)

css = '#snb > div.api_group_option_filter._search_option_simple_wrap > div > div.option_area.type_sort > a:nth-child(2)'
search = driver.find_element_by_css_selector(css)
search.send_keys(Keys.RETURN)
```

* 네이버는 포스팅에 접속할 때 새 탭을 띄우기 때문에 크롤링 시 탭 전환이 이뤄져야한다.

* driver.window_handles는 탭 정보가 담긴 리스트를 반환한다.

* driver.switch_to.window(driver.window_handles[i])로 i번째 탭에 접근할 수 있다.

* driver.close()로 현재 탭을 닫을 수 있다.

```python
for k, wine_list in enumerate(wine_lists[16:]):
    df = DataFrame(columns=['Name', 'Text'])
    
    for j, wine in enumerate(wine_list):
        # 와인 검색
        search = driver.find_element_by_css_selector('#nx_query')
        search.clear()
        search.send_keys(wine)
        search.send_keys(Keys.RETURN)

        # 상위 10개 포스팅
        for i in range(10):
            try:
                search = driver.find_element_by_css_selector(f'#sp_blog_{i + 1} > div > div > a')
                search.send_keys(Keys.RETURN)
            except:
                break

            # 탭 전환
            driver.switch_to.window(driver.window_handles[1])

            # naver blog
            try:
                search = driver.find_element_by_css_selector('#mainFrame')
                new_url = search.get_attribute('src')
                start = new_url.find('logNo')
                end = new_url.find('&', start)
                log = new_url[start + 6:end]
            except:
                driver.close()
                driver.switch_to.window(driver.window_handles[0])
                continue

            driver.get(new_url)
            sleep(1)

            search = driver.find_element_by_css_selector(f'#post-view{log}')
            df.loc[10 * j + i] = [wine, search.text]

            # 탭 닫기
            driver.close()

            # 탭 전환
            driver.switch_to.window(driver.window_handles[0])
        
    # 저장
    df.to_csv(f'wine_dataframe_{k}.csv')
```
